```python
# upload image
uploaded=files.upload()
```

```python
#Load and resize the image
for fn in uploaded.keys():
  image_path=fn
```

```python
img=cv2.imread(image_path)
```

```python
img=cv2.resize(img,(600,600))
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray_blur=cv2.medianBlur(gray,5)
```

```python
edges=cv2.adaptiveThreshold(gray_blur,255,
                            cv2.ADAPTIVE_THRESH_MEAN_C,
                            cv2.THRESH_BINARY,blockSize=9,C=9)
```

```python
#Apply bilateral color for smooth color
color=cv2.bilateralFilter(img,d=9,sigmaColor=300,sigmaSpace=300)
```

```python
# Combine color image with edges
cartoon=cv2.bitwise_and(color,color,mask=edges)
```

```python
#show original and catonnified images
print("original image")
cv2_imshow(img)
```

```python
print("Cartonified image:")
cv2_imshow(cartoon)
```

```python
# save and download the cartoon image
cv2.imwrite("cartoon.jpg",cartoon)
files.download("cartoon.jpg")
```
